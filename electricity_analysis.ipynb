# Electricity Consumption Analysis

This project analyzes electricity usage data to identify consumption patterns and peak demand hours.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/electricity_data.csv')
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['Date'])
```

```python
hourly_consumption = df.groupby(df['Date'].dt.hour)['Global_active_power'].mean()

hourly_consumption.plot(title='Average Hourly Electricity Consumption')
plt.xlabel('Hour of Day')
plt.ylabel('Power Consumption')
plt.show()
```

## Key Insights
- Electricity consumption varies significantly throughout the day.
- Peak usage occurs during specific hours.
- Off-peak periods show lower average power consumption.
