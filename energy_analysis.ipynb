# Energy Demand & Consumption Analytics

This project performs end-to-end analysis of electricity consumption data to identify demand patterns and peak usage periods.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/electricity_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
hourly_demand = df.groupby(df['Date'].dt.hour)['Global_active_power'].mean()

hourly_demand.plot(title='Average Hourly Energy Demand')
plt.xlabel('Hour')
plt.ylabel('Average Power')
plt.show()
```

```python
daily_consumption = df.groupby(df['Date'].dt.date)['Global_active_power'].sum()

daily_consumption.plot(title='Daily Energy Consumption')
plt.xlabel('Date')
plt.ylabel('Total Power')
plt.show()
```

```python
peak_hours = hourly_demand.sort_values(ascending=False).head(3)
off_peak_hours = hourly_demand.sort_values().head(3)

peak_hours, off_peak_hours
```

## Key Insights
- Electricity demand varies significantly across hours of the day
- Peak demand occurs during specific daytime and evening hours
- Off-peak hours show consistently lower consumption
- Daily energy consumption follows a stable pattern
